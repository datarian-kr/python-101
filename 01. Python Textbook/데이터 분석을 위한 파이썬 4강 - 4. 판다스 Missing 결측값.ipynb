결측값을 의미하는 여러 데이터형

#### Na : Not available, 결측값
#### Null : 값이 없다.
#### NaN : Not a Number, 수학적으로 정의되지 않는다.


* 0은 결측값이 아니라 0이라는 값이다

---

### Pandas는 결측치를 np.nan로 대표적으로 표현하고 기본적으로 컴퓨테이션 계산에 포함하지 않는다.

### Reindexing은 특정 열의 인덱스를 바꾸거나 더하거나 삭제하는 것을 하게 하는 메서드.
복사값을 return한다.

#### 데이터에 ios버전 유무 컬럼을 넣어보자

```python
data1 = data.reindex(index=range(5, 15), columns=list(data.columns) +['ios_ver'])
                     
                     
```

그 중 2개만 True값을 넣는다면, 나머지는 NaN값이 뜬다.

```python
data1.loc[0:5:2, 'ios_ver'] = True
```

```python
data1
```

### a. 결측값이 어디라도 있는 데이터를 삭제해보자

```python
data1.dropna(how='any')
```

### b. data.fillna( value = n )

등급 결측치를 삭제하지 말고, 등급의 평균으로 채워보자

데이터 전체를 선택해서 채울수도, 컬럼만 선택해서 채울 수도 있다.

```python
data['등급'].mean().round(1) #round(n)는 소수n자리 수까지 반올림 해준다.
```

```python
data['등급'].fillna(value = data['등급'].mean().round(1))
```

### c. pd.isna(data)

값이 nan인지를 불린boolean으로 보고 싶다면 아래 메서드를 쓰면 된다.

그런데 체크해 보니 fillna로는 데이터가 바뀌지 않았다. copy를 해서 값을 넣어주기 때문!

```python
pd.isna(data)
```

#### 그래서 원본을 수정하고 싶다면 아래처럼 덮어 씌워주면 된다.

```python
data = data['등급'].fillna(value = data['등급'].mean().round(1))
```

```python
pd.isna(data)
```

# 왜 데이터에 결측값이 있으면 안좋을까요? 
### 어떤 상황에 어떤 값으로 채워 넣는 것이 데이터를 최대한 훼손하지 않는 방법일까요?

10명의 사람들의 키 평균을 재려고 하는데, 세 사람이 키를 적어내지 않았다. 이 사람들이 평균에 포함이 될 수 있을까?


판다스 내부 함수들에서 이런 경우들을 자동으로 처리해주는 메서드가 많긴 하지만, 데이터끼리 합치거나 계산하게 될 때 에러가 생기는 경우가 많다.



### a. 내 데이터에 missing 값이 있는지 체크해 보기
- isnull() 메소드는 관측치가 결측이면, True / 결측이 아니면, False의 boollean 값 반환
- notnull()메소드는 관측치가 결측이면, False / 결측이 아니면, True의 boollean 값 반환 -> isnull 정반대.

```python
pd.isnull(data)
```

### b. 칼럼별 결측값 개수 구하기: df.isnull().sum()


boolean 타입은 0(False)과 1(True)로 값이 변환이 되기 때문에 sum()을 할 경우, True의 경우들의 갯수가 나옵니다.

```python
data.isnull().sum()
```

<br>

특정 칼럼 '크기'에 결측값이 몇개 있는지 알고 싶을 때는 컬럼명으로 인덱싱을 해준 후 isnull().sum()을 해주면 됩니다.

```python
data['크기'].isnull().sum()
```

### c. **결측치 연산 방법**

2) mean(), std() 연산 시 : NaN은 분석 대상(population)에서 제외
