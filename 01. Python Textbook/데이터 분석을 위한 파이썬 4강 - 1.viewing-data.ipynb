### e. 정렬해서 보기

### 가장 최근에 업데이트 된 데이터 순으로 보고 싶을 땐 어떻게 할까요? 아니면 리뷰수가 많은 데이터 순으로 보고 싶다면요?

#### .sort_inex()** -> 행과 열 이름 정렬

정렬할 대상 축을 결정할 때, axis.
- axis = 0 => 인덱스 기준 정렬(기본값)
- axis = 1 => 컬럼 기준.

정렬의 방향에 대한 파라미터, ascending.
- ascending = True 오름차순(기본값)
- ascending = False 내림차순

#### 내림차순 정렬 예시

sort_index()를 통해서 인덱스 순서대로, 거꾸로 한번씩 확인해 볼까요?

컬럼을 뒤집어서 볼 수도 있어요.(axis=1)

````python
data.sort_index(axis=0, ascending=True)
````

#### DataFrame 내부 값 정렬
#### 리뷰 수가 많은 순서대로 어플을 정렬해 볼까요?

````python
data.sort_values(by='리뷰 수', ascending = False)
````

#### 설치 수가 많은 순서대로 정렬 해보세요

````python
data.sort_values(by="설치 수", ascending = False)
````
